```python
import numpy as np
import pandas as pd
import seaborn as sns
import seaborn.objects as so
# Apply the default theme
sns.set_theme()
# Load an example dataset
dataTitanic = sns.load_dataset("titanic")
dataTitanic
```

```python
# Objects --> Mark,Stats,Move, dan Scale
# Mark :
  #  Dots Mark
  #  Line Mark
  #  Bar Mark
  #  Fill Mark
  #  Text Mark
# Stats :
  # Agg
  # Est
  # Count
  # Hist
  # KDE
  # Perc
  # PolyFit
# Move :
  # Dodge
  # Jitter
  # Norm
  # Stack
  # Shift
# Scale :
  # Boolean
  # Continuous
  # Nominal
  # Temporal
```

```python
dataTitanic.head()
```

```python
# Mark Object:
# 1. Dots: A mark suitable for dot plots or less-dense scatterplots. Sign cocok digunakan untuk plot titik atau plot sebar yang kurang padat
dataku = so.Plot(dataTitanic, "class", "age")
dataku.add(so.Dots(), color = "alive")

# Kesimpulan :Passenger yang berada pada first class memiliki probability alive lebih tinggi dibandingkan dengan second dan third class.
# Passenger yang ber-usia 0 - 30 tahun memiliki probability alive sangat tinggi dibandingkan dengan passenger yang berusia 30-60 tahun.
```

```python
# Mark Objects:
# 2. Area Objects/ Fill Objects: A fill mark drawn from a baseline to data values.
# Tanda isian/Sign area yang diambil dari garis dasar ke nilai data
p = so.Plot(dataTitanic, "sex", "age").facet("embark_town", wrap=3)
p.add(so.Area())
```

```python
# MARK Objects
# 3. FACET OBJECTS: Produce subplots with conditional subsets of the data.
# Menghasilkan subplots dengan kondisi subset tertentu/himpunan bersyarat pada dataset.
p = so.Plot(dataTitanic, "age", "fare").add(so.Dots())
p.facet("class","sex")
```

```python
# Mark Objects
# 4. Objects Dash: A line mark drawn as an oriented segment for each datapoint. Sign garis yang digambar sebagai segmen berorientasi untuk setiap titik data
# A line segment is drawn for each datapoint, centered on the value along the orientation axis:
p = so.Plot(dataTitanic, "age", "fare")
p.add(so.Dash())
```

```python
# STAT OBJECTS:
# 1. Objects Hist: Class Passenger -> Bin observations, count them, and optionally normalize or cumulate.
so.Plot(dataTitanic, "class").add(so.Bar(), so.Hist())
```

```python
# 2. Objects Count : Embarked Town: Count distinct observations within groups. Menghitung pengamatan yang berbeda di dalam kelompok
so.Plot(dataTitanic, x="embark_town", color="alive").add(so.Bar(), so.Count(), so.Dodge())
```

```python
# STAT OBJECTS:
# 3. OBJECTS AGG: Aggregate data along the value axis using given method.
# Data agregat sepanjang value / sumbu nilai menggunakan metode yang diberikan
p = so.Plot(dataTitanic, "embark_town", "class",color = "alive")
p.add(so.Bar(), so.Agg("mean"))
```

```python
# MOVE OBJECTS:
# 1. Objects Stack: Displacement of overlapping bar or area marks along the value axis.
# Perpindahan bar atau tanda area yang tumpang tindih di sepanjang sumbu nilai.
(
    so.Plot(dataTitanic, x="age", alpha="alive")
    .facet("sex")
    .add(so.Bars(), so.Hist(binwidth=10), so.Stack())
)
```

```python
# MOVE OBJECTS:
# 2. DODGE Objects: Displacement and narrowing of overlapping marks along orientation axis.
# Pemindahan dan penyempitan tanda tumpang tindih sepanjang sumbu orientasi
(
    so.Plot(dataTitanic, "age",color="alive")
    .add(so.Bar(), so.Count(), so.Dodge())
)
```

```python
# Move Objects
# 3. Jitter Objects: Random displacement along one or both axes to reduce overplotting.
# Perpindahan acak sepanjang satu atau kedua sumbu untuk mengurangi overplotting
(
    so.Plot(dataTitanic, "age", "embark_town",color ="alive")
    .add(so.Dots(), so.Jitter())
)
```

```python
# 2. Relative Plots:
# A.) RelPlot: Figure-level interface for drawing relational plots onto a FacetGrid.
# nterface tingkat gambar yang digunakan untuk menggambarkan plot relational ke facetgrid
sns.relplot(data=dataTitanic, x="age", y="fare", hue="alive",col = "sex");
# Kesimpulan nya:
# Gender Male banyak yg tidak selamat. Sedangkan Gender Female banyak yang selamat.
# Gender Male yang selamat berada pada range usia 1-50 tahun.
# Gender Female yang selamat berada pada range 1 - 60 tahun.
```

```python
# B.) Scatter Plot: A scatter plot with possibility of several semantic groupings.
# sebaran plot dengan kemungkinan beberapa pengelompokan semantik
sns.scatterplot(data=dataTitanic, x="class", y = "fare", hue="sex", style = "alive");

# Kesimpulan : Passenger yang berada pada first class dengan sex female memiliki probability alive yang lebih tinggi dibandingkan sex male pada first class.
# Female memiliki probability alive lebih tinggi dibandingkan male.
# First class merupakan class paling aman pada Titanic, karena passenger yang berada pada first class paling banyak selamat.
```

```python
# C.) Line Plot: A line plot with possibility of several semantic groupings.
# plot garis dengan kemungkinan beberapa pengelompokan semantik.
sns.lineplot(data=dataTitanic, x="pclass", y="fare", hue="sex",style = "alive",markers=True, dashes=False);
```
