```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
dataTitanic = sns.load_dataset("titanic")
```

```python
# 1. Objects:
# A. Plot:
import seaborn as sns
import seaborn.objects as so
sns.set_theme(style="ticks")

# Show the joint distribution using kernel density estimation
g = sns.jointplot(
    data=dataTitanic,
    x="age", y="fare", hue="alive",
    kind="kde",
)
```

```python
# Mark Object:
# 1. Dots:
dataku = so.Plot(dataTitanic, "class", "age")
dataku.add(so.Dots(), color = "alive")
```

```python
# Mark Objects:
# 2. Area Objects/ Fill Objects:
p = so.Plot(dataTitanic, "sex", "age").facet("embark_town", wrap=3)
p.add(so.Area())
```

```python
# MARK Objects
# 3. FACET OBJECTS:
p = so.Plot(dataTitanic, "age", "fare").add(so.Dots())
p.facet("class","sex")
```

```python
# 4. Objects Dash:
p = so.Plot(dataTitanic, "age", "fare")
p.add(so.Dash())
```

```python
p = so.Plot(dataTitanic, "fare", "agew", color="alive")
p.add(so.Path())
```

```python
# STAT OBJECTS:
# 1. Objects Hist: Class Passenger
so.Plot(dataTitanic, "class").add(so.Bar(), so.Hist())
```

```python
# 2. Objects Count : Embarked Town:
so.Plot(dataTitanic, x="embark_town", color="alive").add(so.Bar(), so.Count(), so.Dodge())
```

```python
# STAT OBJECTS:
# 3. OBJECTS AGG:
p = so.Plot(dataTitanic, "embark_town", "class")
p.add(so.Bar(), so.Agg("mean"))
```

```python
# MOVE OBJECTS:
# 1. Objects Stack:
(
    so.Plot(dataTitanic, x="age", alpha="alive")
    .facet("sex")
    .add(so.Bars(), so.Hist(binwidth=10), so.Stack())
)
```

```python
# MOVE OBJECTS:
# 2. DODGE Objects:
(
    so.Plot(dataTitanic, "age",color="alive")
    .add(so.Bar(), so.Count(), so.Dodge())
)
```

```python
# Move Objects
# 3. Jitter Objects:
(
    so.Plot(dataTitanic, "age", "embark_town",color ="alive")
    .add(so.Dots(), so.Jitter())
)
```

```python
# Relative Plots:
# RelPlot:
sns.relplot(data=dataTitanic, x="age", y="fare", hue="alive");
```

```python
# 2. Relative Plots:
# A.) RelPlot + Column/Time parameter:
sns.relplot(data=dataTitanic, x="age", y="fare", hue="alive",col = "sex");
# Kesimpulan nya:
# Gender Male banyak yg tidak selamat. Sedangkan Gender Female banyak yang selamat.
# Gender Male yang selamat berada pada range usia 1-50 tahun.
# Gender Female yang selamat berada pada range 1 - 60 tahun.
```

```python
# B.) Scatter Plot:
sns.scatterplot(data=dataTitanic, x="class", y = "fare", hue="sex", style = "alive");
```

```python
# Scatter Plot with Size:
sns.scatterplot(data=dataTitanic, x="age", y="fare", size="pclass",hue = "alive");
```

```python
# RelPlot Combine with Scatter Plot:
sns.relplot(
    data=dataTitanic, x="age", y="fare",
    col="class", hue="alive", style="sex",
    kind="scatter"
);
```

```python
# C.) Line Plot:
sns.lineplot(data=dataTitanic, x="pclass", y="fare", hue="sex",style = "alive",markers=True, dashes=False);
```
